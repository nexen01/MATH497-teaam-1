```python
from geopy.geocoders import Nominatim
def extract_location(addr):
  loc = Nominatim(user_agent="GetLoc")
  try:
    getLoc = loc.geocode(addr + " Chicago, IL")
    return [getLoc.latitude, getLoc.longitude]
  except:
    return [0,0] 
```

```python
def unanonymize_address(addr):
  addr = str(addr)
  if(addr == "nan" or addr == "0  UNKNOWN"): #no address to anonymize 
    return addr

  tokens = addr.split(" ")
  tokens[0] = tokens[0][:-2] + "00"
  return " ".join(tokens)
unanonymize_address("32XX S STATE ST")
```

```python
leadresults['numAddr'] = leadresults['Address'].apply(unanonymize_address)
```
